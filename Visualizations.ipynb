```r
data("midwest", package = "ggplot2")
```

```r
head(midwest)
```

```r
str(midwest)
options(scipen=999)  # turn-off scientific notation like 1e+48
```

```r
ggplot(data = midwest, mapping=aes(x=area, y=poptotal)) + 
geom_point(mapping = aes(color=state, size = popdensity)) +
#scale_y_log10() + 
xlim(c(0, 0.1)) + 
ylim(c(0, 500000)) +
labs(y = 'population', x = 'area', 
     subtitle = 'population vs area scatterplot', title="Scatterplot", caption = "Source: midwest") + 
geom_smooth(se = F)
```

```r
# Mark certain areas
library(ggalt)
markArea <- midwest[midwest$poptotal > 350000,]
```

```r
ggplot(data = midwest, mapping=aes(x=area, y=poptotal)) + 
geom_point(mapping = aes(color=state, size = popdensity)) +
#scale_y_log10() + 
xlim(c(0, 0.1)) + 
ylim(c(0, 500000)) +
labs(y = 'population', x = 'area', 
     subtitle = 'population vs area scatterplot', title="Scatterplot", caption = "Source: midwest") + 
geom_smooth(se = F) +
geom_encircle(data = markArea, mapping = (aes(x=area, y=poptotal)), color = "RED", size = 2)
```

```r
data(mpg, package="ggplot2")
head(mpg)
```

```r
# Jitter to uncluster overlapping points

ggplot(data = mpg, mapping = aes(cty, hwy)) + 
geom_point(mapping = aes(color = class)) + 
geom_smooth(se=F) +
labs(x="city mileage", y = "highway mileage", title = "city vs highway scatterplot jitter") +
geom_jitter(size = 1 , width = 0.5)
```

```r
# Other option of jitter is count charts. But too many overlapping points can cause issues.

ggplot(data = mpg, mapping = aes(cty, hwy)) + 
geom_count(color="RED", show.legend = F) +
geom_smooth(se=F) +
labs(x="city mileage", y = "highway mileage", title = "city vs highway scatterplot count") 
```

```r
head(mpg)
```

```r
# Bubble plot
mpg_select <- mpg[mpg$manufacturer %in% c("audi", "ford", "honda", "hyundai"), ]

ggplot(data = mpg_select, mapping = aes(x = displ, y = cty)) + 
geom_jitter(mapping = aes(col = manufacturer, size = hwy)) + 
labs(x = "displacement",
     y = "city mileage",
     title = "bubble plot",
     subtitle = "displacement vs city mileage",
     caption = "mpg") +
geom_smooth(se = F, method="lm", mapping = aes(col = manufacturer))
```

```r
# animated bubble chart
library(gganimate)
library(gapminder)
head(gapminder)
```

```r
g <- ggplot(data=gapminder, mapping = aes(x=gdpPercap, y=lifeExp)) + 
geom_point(mapping = aes(col = continent, size=pop, frame = year)) +
labs(x="GDP Per capita",
     y="life expectancy",
     title="animated bubble",
     subtitle="GDP vs life expectancy",
     caption="gapminder") +
  facet_wrap(~continent) 
gganimate(g, interval=0.2)
```
